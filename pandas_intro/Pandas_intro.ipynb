<center><h1>HANDLING OUTLIERS</h1></center>

```python
##Finding the 1st Quartile (Q1) and 3rd Quartile (Q3) of Defense column

Q1 = df['Defense'].quantile(0.25)  # 25th percentile (Lower Quartile)
Q3 = df['Defense'].quantile(0.75)  # 75th percentile (Upper Quartile)

print(Q1, "- LOWER QUARTILE"), print(Q3, "- UPPER QUARTILE")

##CALCULATING THE INTERQUATILE RANGE
IQR = Q3 - Q1
print(IQR,"- INTERQUATILE RANGE")

##CALCULATING THE OUTLIER BOUNDARIES
lower_bound = Q1 - 1.5 * IQR  # Lower outlier threshold
upper_bound = Q3 + 1.5 * IQR  # Upper outlier threshold

print(lower_bound, "- Lower outlier threshold"), print(upper_bound,"-Upper outlier threshold")

##DETECTING OUTLIERS
outliers = df[(df['Defense'] < lower_bound) | (df['Defense'] > upper_bound)]
print(outliers)
```
